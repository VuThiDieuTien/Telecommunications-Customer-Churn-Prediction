```python
import pandas as pd
df= pd.read_excel(r'/content/clustering.xlsx')
```

```python
contract_map = {
    "Month-to-Month": 0,
    "One Year": 1,
    "Two Year": 2
}

df["contract_length"] = df["contract"].map(contract_map)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import numpy as np

features = [
    "tenure_in_months",
    "monthly_charge",
    "number_of_referrals",
    "number_of_dependents",
    "contract_length"
]

X = df[features].copy()

# bỏ giá trị âm vô nghĩa (vd monthly_charge = -10)
X[X < 0] = np.nan

# fill missing
X = SimpleImputer(strategy="median").fit_transform(X)

# scale để các biến có trọng số công bằng
X_scaled = StandardScaler().fit_transform(X)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

df["cluster"] = kmeans.fit_predict(X_scaled)
```

```python
import pandas as pd

churn_rate = pd.crosstab(
    df["cluster"],
    df["customer_status"],
    normalize="index"
) * 100

churn_rate
```
